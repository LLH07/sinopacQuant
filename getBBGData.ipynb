```python
import pandas as pd
import numpy as np
from xbbg import blp
```

```python
startDate = '2012-01-01'
endDate = '2025-04-18'

dr = pd.date_range(start=startDate, end=endDate, freq='B')
basket = ['EUR', 'JPY', 'GBP', 'CHF', 'AUD', 'CAD', 'NZD', 'SEK', 'ZAR', 'CNH']
spot = pd.DataFrame(index = dr, columns=basket)
```

```python
for i in basket:
    print(f"Fetching data for {i}...")
    spot[i] = blp.bdh(tickers=f'{i} CMTK Curncy', flds = ["PX_LAST"], start_date=startDate, end_date=endDate).reindex(dr).ffill()
```

```python
spot.to_csv('@@spot.csv')
```

```python
## for BGNT data
openBGNT = pd.DataFrame(index = dr, columns=basket)
closeBGNT = pd.DataFrame(index = dr, columns=basket)

for i in basket:
    print(f"Fetching data for {i}...")
    openBGNT[i] = blp.bdh(tickers=f'{i} BGNT Curncy', flds = ["PX_OPEN"], start_date=startDate, end_date=endDate).reindex(dr).ffill()
    closeBGNT[i] = blp.bdh(tickers=f'{i} BGNT Curncy', flds = ["PX_LAST"], start_date=startDate, end_date=endDate).reindex(dr).ffill()
```

```python
openBGNT.to_csv('@@openBGNT.csv')
closeBGNT.to_csv('@@closeBGNT.csv')
```
